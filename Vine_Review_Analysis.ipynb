```python
import pandas as pd
```

```python
vine_df = pd.read_csv("./Data/vine_table.csv")
vine_df
```

```python
vine_df.info()
```

```python
vine_df_20 = vine_df.loc[vine_df["total_votes"] >= 20]
vine_df_20
```

```python
vine_df_50 = vine_df_20.loc[(vine_df_20["helpful_votes"] / vine_df_20["total_votes"]) >= 0.50]
vine_df_50
```

```python
vine_paid_review = vine_df_50.loc[vine_df_50["vine"] == 'Y']
vine_paid_review
```

```python
vine_unpaid_review = vine_df_50.loc[vine_df_50["vine"] == 'N']
vine_unpaid_review
```

```python
paid_star_reviews = vine_paid_review.loc[vine_paid_review["star_rating"] == 5]["star_rating"].count()
paid_star_reviews
```

```python
paid_reviews = vine_paid_review["star_rating"].count()
paid_reviews
```

```python
percentage_paid_star_reviews = round((paid_star_reviews / paid_reviews) * 100, 2)
percentage_paid_star_reviews
```

```python
vine_star_unpaid_review = vine_unpaid_review.loc[vine_unpaid_review["star_rating"] == 5]["star_rating"].count()
vine_star_unpaid_review 
```

```python
unpaid_reviews = vine_unpaid_review["star_rating"].count()
unpaid_reviews
```

```python
percentage_unpaid_star_reviews = round((vine_star_unpaid_review  / unpaid_reviews) * 100, 2)
percentage_unpaid_star_reviews
```
